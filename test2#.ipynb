```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread("3.jpg")

h, w = img.shape[:2] 
mask = np.zeros((h+2, w+2), np.uint8)
cv2.floodFill(img, mask, (w-1,h-1), (255,255,255), (2,2,2),(3,3,3),8)
```

```python
#圖像灰度化
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python

blur=cv2.GaussianBlur(gray,(29,29),1,1)
cv2_imshow(blur)
```

```python


import cv2
import numpy as np
#np.set_printoptions(threshold=np.inf)  
import matplotlib.pyplot as plt
blur=~blur
kernel = np.ones((15,15), np.uint8)

img1=cv2.erode(blur,kernel,30)

titles = ['Original Image', 'Erosion(15x15,30)']  
images = [blur, img1] 
for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i],"gray")  
    plt.title(titles[i])    
    plt.axis('off')
plt.show()
```

```python
ret,thresh1=cv2.threshold(img1,6,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
cv2.waitKey(0)
```

```python
kernel = np.ones((7,7), np.uint8)
close=cv2.morphologyEx(thresh1, cv2.MORPH_CLOSE, kernel)
cv2_imshow(close)
cv2.waitKey(0)
```

```python

Canny = cv2.Canny(close, 50, 150)

cv2_imshow(Canny)

cv2.waitKey(0)
```

```python

(cnts1,_) =cv2.findContours(Canny,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
img2 = img.copy()
cv2.drawContours(img2, cnts1, -1, (0,255,0), 2)

cv2_imshow(img2)

cv2.waitKey(0)
```
